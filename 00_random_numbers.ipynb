## How random numbers are generated on computers?

A **Linear Congruential Generator (LCG)** is a simple and widely used algorithm for generating pseudo-random numbers. It works by applying a recursive formula to produce a sequence of numbers, with the seed or starting value being the first element of the sequence. The resulting sequence, while not truly random, exhibits statistical properties similar to those of a random sequence, making it suitable for many applications. 

`Xn+1 = (a * Xn + b) mod m`

where
Xn: is the current number in the sequence.

Xn+1: is the next number in the sequence.

a: is the multiplier (a constant).

b: is the increment (a constant).

m: is the modulus (a constant). 


````python
def lcg(a, c, m, seed, n):
    """
    Generate n values of the LCG x_{k+1} = (a x_k + c) mod m.
    """
    x = seed
    seq = []
    for _ in range(n):
        seq.append(x)
        x = (a * x + c) % m
    return seq

# Parameters for our tiny LCG
a, c, m, seed = 13, 7, 100, 1
n = 200

# Generate the sequence
seq = lcg(a, c, m, seed, n)

# Plot the values vs. step index
plt.figure(figsize=(8,4))
plt.plot(range(n), seq, marker='o', linestyle='-')
plt.title(rf"LCG Sequence (a={a}, c={c}, m={m}) repeats every $\le${m:d} steps")
plt.xlabel("Iteration")
plt.ylabel("X_k value")
plt.show()
````

Let us compare this with theoretical gaussian distribution.

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

````python
# defining a gaussian function

def gaussian(mu, sigma):
    return 1/np.sqrt(2*np.pi*sigma**2) * np.exp(-(x-mu)**2/(2*sigma**2))
````

````python
x = np.linspace(-40, 40)
fx = gaussian(mean, std)

plt.plot(x, fx)
plt.axvline(mean, ls='--')
plt.show()
````

````python
x = np.linspace(-40, 40, 500)
fx = gaussian(mean, std)

plt.plot(x, fx)
plt.axvline(mean, ls='--')
plt.show()
````
