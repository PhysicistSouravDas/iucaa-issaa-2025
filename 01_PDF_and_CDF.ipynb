# PDF and CDF

Probability Distribution Functions and Cumulative Distribution Function

If $P(x)$ is a PDF, then

$$
{\rm CDF} = \int_0^x P(x)dx
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import cumulative_trapezoid as cumtrapz

x = np.linspace(-3, 3, 500)
pdf = norm.pdf(x)
cdf = cumtrapz(pdf, x, initial=0)

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density / Probability')
plt.show()
```

```python
# Doing with our gaussian function

def gaussian(mu, sigma):
    return 1/np.sqrt(2*np.pi*sigma**2) * np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
x = np.linspace(-5, 5, 500)
pdf = gaussian(1, 2)
cdf = cumtrapz(pdf, x, initial=0)

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density / Probability')
plt.show()
```

```python
x = np.linspace(-5, 5, 500)
pdf = gaussian(0, 0.5)
cdf = cumtrapz(pdf, x, initial=0)

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density / Probability')
plt.show()
```

```python
# uniform PDF

x = np.linspace(1, 5, 500)
pdf = 0.0*x
pdf[:] = 1/(5-1)
cdf = cumtrapz(pdf, x, initial=0)

plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='PDF')
plt.plot(x, cdf, label='CDF')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density / Probability')
plt.show()
```
